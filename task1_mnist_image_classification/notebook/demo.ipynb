### MNIST Classifier Demo

```python
import sys
import os

sys.path.append(os.path.abspath(".."))
```

```python
from  models.mnist_classifier import MnistClassifier
from  utils.mnist_loader import load_data
from sklearn.metrics import accuracy_score
```

```python
# load dataset
X_train, y_train, X_test, y_test = load_data()
```

```python
# Random forest
model = MnistClassifier("rf")

model.train(X_train, y_train)

preds = model.predict(X_test)

print("RandomForest Accuracy:", accuracy_score(y_test, preds))
```

```python
# FFNN
model = MnistClassifier("nn")

model.train(X_train, y_train)

preds = model.predict(X_test)

print("Neural Network Accuracy:", accuracy_score(y_test, preds))
```

```python
#CNN
model = MnistClassifier("cnn")

model.train(X_train, y_train)

preds = model.predict(X_test)

print("CNN Accuracy:", accuracy_score(y_test, preds))
```

```python
# Edge Case 1: correct single image prediction

single_image = X_test[5].reshape(1,28,28)
prediction = model.predict(single_image)
print("✅ Edge Case 1: Single image prediction (correct input)")
print("Prediction:", prediction)
```

```python
# Edge Case 2: Incorrect input shape (missing batch dimension)
wrong_image1 = X_test[5] 
try:
    pred = model.predict(wrong_image1)
except Exception as e:
    print("❌ Edge Case 2: Missing batch dimension")
    print("Error:", e)
```

```python
# Edge Case 3: Truly incorrect shape
wrong_image2 = X_test[5].reshape(1, 14, 56)  # 14*56 != 28*28
try:
    pred = model.predict(wrong_image2)
except Exception as e:
    print("❌ Edge Case 3: Incorrect shape (14x56)")
    print("Error:", e)
```
