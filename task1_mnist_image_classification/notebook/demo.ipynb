### MNIST Classifier Demo

```python
import sys
import os
print(os.getcwd())

sys.path.append(os.path.abspath(".."))
```

```python
from  models.mnist_classifier import MnistClassifier
from  utils.mnist_loader import load_data
from sklearn.metrics import accuracy_score
```

```python
# load mnist
X_train, y_train, X_test, y_test = load_data()
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
# train model
model = MnistClassifier(algorithm="rf")

model.train(X_train, y_train)
```

```python
#predict
preds = model.predict(X_test)
#evaluate
print("Accuracy:", accuracy_score(y_test, preds))
```

```python
import matplotlib.pyplot as plt
plt.imshow(X_test[0], cmap="gray")
plt.title(f"Label: {y_test[0]} | Prediction: {preds[0]}")
plt.show()
```

```python
# Edge Case 1: Single image prediction

single_image = X_test[0]

single_image = single_image.reshape(1,28,28)

prediction = model.predict(single_image)

print(prediction)
```

```python
# Edge Case 2 incorrect input shape
wrong_image = X_test[0].reshape(14,56)

model.predict(wrong_image)
```

```python
# Edge Case 2 Empty input
import numpy as np

empty = np.array([])

model.predict(empty)
```
