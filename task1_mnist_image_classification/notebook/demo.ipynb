### MNIST Classifier Demo

This notebook demonstrates the usage of the implemented classifiers:
- Random Forest
- Feed-Forward Neural Network
- Convolutional Neural Network

It also includes several edge cases to verify model robustness.

```python
import sys
import os
project_root = os.path.abspath("..")
if project_root not in sys.path:
    sys.path.append(project_root)
```

```python
from  models.mnist_classifier import MnistClassifier
from  utils.mnist_loader import load_data
from sklearn.metrics import accuracy_score
```

```python
# load dataset
X_train, y_train, X_test, y_test = load_data()

# reduce dataset for faster demo
X_train = X_train[:10000]
y_train = y_train[:10000]

X_test = X_test[:2000]
y_test = y_test[:2000]

# train models
algorithms = ["rf", "nn", "cnn"]

for algo in algorithms:
    print(f"\nTraining {algo.upper()} model")

    model = MnistClassifier(algorithm=algo)
    model.train(X_train, y_train)

    preds = model.predict(X_test)
    acc = accuracy_score(y_test, preds)

    print(f"{algo.upper()} Accuracy: {acc:.4f}")
```

```python
# Edge Case 1: Single image prediction 

single_image = X_test[5].reshape(1,28,28)
prediction = model.predict(single_image)
print("✅ Edge Case 1: Single image prediction")
print("Predicted label:", prediction[0])
print("Actual label:", y_test[5])
```

```python
# Edge Case 2: Incorrect input shape (missing batch dimension)
wrong_image1 = X_test[5] 
try:
    pred = model.predict(wrong_image1)
except Exception as e:
    print("❌ Edge Case 2: Missing batch dimension")
    print("Error:", e)
```
